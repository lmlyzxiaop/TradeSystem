```python
#2017 2016 2015 2014 2013 2012 2011 2010

#7年收益率变化

x = [2011, 2012, 2013, 2014, 2015, 2016, 2017]

y = [0.12, 0.06, 0.41, 0.21, 0.09, 0.67, 2.71]

plt.xlabel(r"year")

plt.ylabel(r'rate of return')

plt.title('BoatWater')

plt.legend(loc = 'upper left')

plt.plot(x,y, label = "Rate of Return")  
```

```python
#资产分布
plt.figure(figsize=(6,6));

x = [5, 10, 10, 5, 30, 40]

labels = ['Convertible Bond', ' Graded Fund', 'ETF Fund', 'P2P', 'Shares', 'Commodity Futures'] 

explode = [0.2, 0.1, 0, 0, 0.1, 0] 

plt.pie(x, labels=labels, explode=explode, autopct='%1.1f%%')

plt.title("BoatWater")

plt.show()
```
